```python
from pydal import DAL ,Field
```

```python
#the syntax to connect with database 
dbConnect = DAL('mysql://root:root@localhost/studyCode')

try:
    #The syntax to create students table and insert data in it
    dbConnect.define_table('students', Field('name'), Field('age', type = 'integer'))
    #dbConnect.students.insert(name = 'Nguyen Van A', age = 21)
    #dbConnect.students.insert(name = 'Nguyen Van B', age = 23)
    
    #use function to select all rows in students table,we can implement join syntax on select function
    getAllStudents = dbConnect().select(dbConnect.students.ALL)
    
    #get all students in student table
    print("get All students")
    for student in getAllStudents:
        print(student['name'],student['age'])
    
   #use sort function to order student
    print("The list of students have been sorted")
    for sortStudent in getAllStudents.sort(lambda sortStudent: sortStudent.name):
        print(sortStudent.name)
        
    #Use find function to find the name of students have chacracter 'N' in it's name 
    print("The list of students have been Finded")
    for findStudent in getAllStudents.find(lambda findStudent: findStudent.name[0] == 'N'):
        print(findStudent.name)
    
    #Use exclude to return students have chacracter 'N' in it's name
    print("Exclude student")
    for excludeStudent in getAllStudents.exclude(lambda excludeStudent: excludeStudent.name[0] == 'P'):
        print(excludeStudent.name)
           
    #select all rows in students table and order by age
    getOrderByStudent = dbConnect(dbConnect.students).select(orderby = dbConnect.students.age)
    
    #get all students in student table after order by it
    print("Get all students have been ordered")
    for orderStudent in getOrderByStudent:
        print(orderStudent['name'],orderStudent['age'])
        
    #select limit data in students table
    getLimitStudent = dbConnect(dbConnect.students).select(limitby=(1,3))
    
    #get all students in student table after order by it
    print("Get limit student")
    for orderStudent in getLimitStudent:
        print(orderStudent['name'],orderStudent['age'])
        
    #get count int table
    countStudents = dbConnect(dbConnect.students.id).count()
    
    #Output number of student in school
    print("The total students is :%r"%countStudents)
    
    #use json to return values
    showStudents = dbConnect(dbConnect.students).select()
    
    #print value return from json
    print(showStudents.as_json())
    
    #The method used query statement to select student based on id
    def select_query(param):
        selQuery = "select * from students where id = %r" % param
        return selQuery
    
    #return value bafore execute select syntax
    getStudentById = dbConnect.executesql(select_query(1))
    
    #print return value
    print(getStudentById)
    
    #the method used query statement to delete student based on id
    def delete_query(studentId):
        delQuery = "delete from students where id = %r" %studentId
        return delQuery
    
    #execute delete query
    print("Remove student")
    dbConnect.executesql(delete_query(2))
    
    #the medthod used query statement to update student
    def update_query(name,age,stuId):
        upQuery = "update students set name = '%s' ,age = %r where id = %r" %(name,age,stuId)
        return upQuery
    
    #execute update query
    print("Execute Update Student")
    dbConnect.executesql(update_query('Nals solution',5,4))
    print("Success!")
    
    #The method used query statement to insert data in table
    def insert_query(name,age):
        insertQuery = "insert into students (name,age) values('%s',%r)"%(name,age)
        return insertQuery
    
    #execute insert syntax
    dbConnect.executesql(insert_query('Python',23))
    
    print("Insert success!")
    
    #Use function to implement delete,update
    
    #==>use function to delete records in table
    print("Please! Input the student you want to delete")
    inputDelete = input()
    dbConnect(dbConnect.students.id == inputDelete).delete()
    print("Delete success")
    
    #==>Use function to update record in table
    print("Please! Input the student you want to update")
    inputStudent = input()
    dbConnect(dbConnect.students.id == inputStudent).update(name = 'NAL Solution', age = 30)
    
    #use update_record function
    updateStudent = dbConnect(dbConnect.students.id == inputStudent).select().first()
    updateStudent.update_record(name = 'Simulater' ,age = 55)
    print("Update success!")
    
    # The sample get on website:http://www.web2py.com/books/default/chapter/29/06/the-database-abstraction-layer#callbacks-on-record-insert-delete-and-update

finally:
    if dbConnect:
        dbConnect.close()
```
